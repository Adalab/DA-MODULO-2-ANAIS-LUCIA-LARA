2. Cread dos dataframes nuevos, uno que incluya solo las variables categóricas y otro que incluya solo las variables numéricas.

- Extraed los valores únicos su frecuencia. BONUS presenta los resultados en un dataframe. Os deberá quedar algo como esto:

```python
lista_objects = []
lista_floats = []

for i in df.columns:
    if df[i].dtypes == "object":
        lista_objects.append(i)
    else:
        lista_floats.append(i)
```

```python
df_objects = df[lista_objects]
df_objects
```

```python
df_objects.describe().T
```

```python
df_floats = df[lista_floats]
df_floats
```

```python
df_floats.describe().T
```

```python
unicos_floats= df_floats["year"].unique()
unicos_floats
```

```python
years= df.groupby("year")
```

```python
years2= years.agg(frecuencia= ("year", "count")).reset_index()
years2
```
